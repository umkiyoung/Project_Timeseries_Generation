```python
import matplotlib.pyplot as plt
from dataloader import DataloaderHandler

# load dataloader handler
dlh = DataloaderHandler(batch_size=128,
                        shuffle=True,
                        drop_last=True)
```

```python
# sine dataloader
sine_dl = dlh.sine_dataloader(sample_n=10000,
                              seq_len=24,
                              feature_n=5)
X = next(iter(sine_dl))
print(X)
print(X.shape)

# visualize
plt.figure()
plt.plot(X[0])
plt.show()
```

```python
# synthetic time-series data
syn_dl = dlh.synthetic_dataloader(sample_n=10000,
                                  seq_len=24,
                                  freq=1,
                                  noise_level=0.1,
                                  trend=True)

X = next(iter(syn_dl))
print(X)
print(X.shape)

# visualize
plt.figure()
plt.plot(X[0])
plt.show()
```

```python
# stock real data
stock_dl = dlh.stock_dataloader(symbols=["AAPL", "GOOG"],
                                sdate="2010",
                                window=24)

X = next(iter(stock_dl))
print(X)
print(X.shape)

# visualize
plt.figure()
plt.plot(X[0])
plt.show()
```
